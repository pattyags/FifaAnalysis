## Import matplotlib to start creating visualizations ##

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

### Start with Histograms ###

Visualizing the distribution of birth years among players

```python
figure, axes = plt.subplots()
axes.hist(players_15_df['dob'])
axes.legend()
axes.set_title('Player Date of Birth Distribution in 2015')
axes.set_xlabel('Year')
axes.set_ylabel('Count of Players')
```

Visualizing the distribution of ages among players

```python
figure, axes = plt.subplots()
axes.hist(players_15_df['age'])
axes.legend()
axes.set_title('Player Age Distribution in 2015')
axes.set_xlabel('Age')
axes.set_ylabel('Count of Players')
```

```python
figure, axes = plt.subplots()
axes.hist(players_15_df['height_cm'])
axes.legend()
axes.set_title('Player Height Distribution in 2015')
axes.set_xlabel('Height (in cm)')
axes.set_ylabel('Count of Players')
```

```python
figure, axes = plt.subplots()
axes.hist(players_15_df['weight_kg'])
axes.legend()
axes.set_title('Player Weight Distribution in 2015')
axes.set_xlabel('Weight (in kg)')
axes.set_ylabel('Count of Players')
```

```python
players_by_nationality = players_15_df.groupby(["nationality"])

count_nationalities = players_by_nationality['nationality'].count()
count_nationalities.head()
```

```python
count_nationalities.sort_values(ascending = False, inplace=True)
count_nationalities.head()
```

```python
count_nationalities.sort_values(ascending = False, inplace=True)
count_nationalities.head()
```

```python
count_top15_nationalities = count_nationalities[0:15]
count_top15_nationalities
```

```python
count_restof_nationalities = count_nationalities[15:]
count_restof_nationalities.head()
```

```python
add_other_column = count_restof_nationalities.sum()
```

```python
count_top15_nationalities['Other'] = add_other_column
count_top15_nationalities
```

```python
figure, axes = plt.subplots()

count_top15_nationalities.plot(ax=axes, kind='bar', color='orange')
```

```python
only_top_10_nat = count_top15_nationalities.drop('Other')[0:10]
only_top_10_nat
```

```python
plt.pie(only_top_10_nat.values, labels=only_top_10_nat.index, autopct='%1.1f%%')
```
